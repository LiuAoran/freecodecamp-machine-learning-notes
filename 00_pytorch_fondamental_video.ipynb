## Matrix multiplication

Two main ways of performing multiplication in neural networks and deep learning:

1. Element-wise multiplication.
   
![alt text](source/matrix-multiply-constant.svg)

2. Matrix multiplication (dot product).
   
![alt text](source/matrix-multiply-a.svg)

![alt text](source/matrix-multiply-c.svg)

More information on multiplying matrices. - https://www.mathsisfun.com/algebra/matrix-multiplying.html

There are two main rules that performing matrix multiplication needs to satisfy:
1. The *inner dimensions* must match:
* `(3, 2) @ (3, 2)` won't work
* `(2, 3) @ (3, 2)` will work
* `(3, 2) @ (2, 3)` will work
2. The resulting matrix has the shape of the *outer dimensions*:
* `(2, 3) @ (3, 2) -> (2, 2)`
* `(3, 2) @ (2, 3) -> (3, 3)`

```python
# torch.matmul(torch.rand(3,2), torch.rand(3,2)) won't work because the inner dimensions don't match. 
torch.matmul(torch.rand(3,2), torch.rand(2,3)) # works because the inner dimensions match (2 and 2) and the output shape is determined by the outer dimensions (3 and 3) -> torch.Size([3, 3])
```

## One of the most common errors in deep learning: shape errors

```python
# Shapes for matrix multiplication
tensor_A = torch.tensor([[1, 2], 
                          [3, 4], 
                          [5, 6]])
tensor_B = torch.tensor([[7, 10],
                         [8, 11],
                         [9, 12]])
# torch.mm(tensor_A, tensor_B) # torch.mm() is an alias for torch.matmul(). (It is an alias for writing less code.)
torch.matmul(tensor_A, tensor_B) # same as torch.mm() when both inputs are 2D tensors
```

```python
tensor_A.shape, tensor_B.shape
```

To fix our tensor shape issues, we can manipulate the shape of one of our tensors using a **transpose**.

A **transpose** switches the axes or dimensions of a giving tensor.

```python
tensor_B, tensor_B.shape
```

```python
tensor_B.T,tensor_B.T.shape
```

```python
# The matrix multiplication operation works when tensor_B is transposed 
# because the inner dimensions match (2 and 2) 
# and the output shape is determined by the outer dimensions (3 and 3) -> torch.Size([3, 3])

print(f"Original shapes: tensor_A: {tensor_A.shape}, tensor_B: {tensor_B.shape}")
print(f"New shapes: tensor_A: {tensor_A.shape}(same shape as above), tensor_B.T: {tensor_B.T.shape}")
print(f"Multiplying: {tensor_A.shape} @ {tensor_B.T.shape} <- inner dimensions must match")
print("Output: \n")
output = torch.matmul(tensor_A, tensor_B.T)
print(output)
print(f"\nOutput shape: {output.shape} <- determined by the outer dimensions of the original shapes of tensor_A and tensor_B.T")
```

## Reshaping, stacking, squeezing and unsqueezing tensors

* Reshaping - reshapes an input tensor to a defined shape.
* View - Return a view of an input tensor of certain shape but keep the same memory as the original tensor.
* Stacking - combine multiple tensors on top of each other (vstack) or side by side (hstack).
* Squeeze - removes all `1` dimensions from a tensor.
* Unsqueeze - add a `1` dimension to a target tensor.
* Permute - Return a view of the input with dimensions permuted(swapped) in a certain way.

### Reshape and view

**Note:**
Reshape may copy or rearrange data to fit a new shape, while view returns a new tensor sharing the same memory and requires the tensor to be contiguous.


```python
# Let's create a tensor
import torch
x = torch.arange(24.)
x, x.shape
```

```python
# Add an extra dimension
x_reshaped = x.reshape(1, 24)    
x_reshaped, x_reshaped.shape
```

```python
# Add 3D dimension
x_reshaped_3d = x.reshape(2, 3, 4)
x_reshaped_3d, x_reshaped_3d.shape
```

```python
# Change the view
z = x.view(1, 24)
z, z.shape
```

```python
# Changing z changes x 
# (because a view of a tensor shares the same memory as the original input)

z[:, 0] = 5
z, x
```

### Stack

dim tells which layer the stacked tensors go into.

```python
x = torch.arange(6.)
x, x.shape
```

```python
# Stack tensors on top of each other
x_stacked = torch.stack([x, x, x, x], dim=0)
print(f"x_stacked while dim = 0: \n {x_stacked}")
print(f"x_stacked.shape: {x_stacked.shape} \n")
x_stacked_dim1 = torch.stack([x, x, x, x], dim=1) 
print(f"x_stacked while dim = 1: \n {x_stacked_dim1}")
print(f"x_stacked_dim1.shape: {x_stacked_dim1.shape} \n")
```

```python
x = torch.tensor([[1, 2],
                  [3, 4]])  # shape = [2, 2]
x_stacked = torch.stack([x, x, x], dim=1)
x_stacked, x_stacked.shape
```

### Squeeze && UnSqueeze

**Squeeze:** removes all single dimensions from a target tensor.

**Unsqueeze:** adds a single dimension to a target tensor at a specific dim(dimension).

```python
x = torch.arange(1., 7.)
x_reshaped = x.reshape(1, 6)
print(f"Previous tensor: \n {x_reshaped}")
print(f"Previous shape: {x_reshaped.shape} \n")

#Remove all the extra dimensions with squeeze()
x_squeezed = x_reshaped.squeeze()
print(f"New tensor: \n {x_squeezed}")
print(f"New shape: {x_squeezed.shape}")
```

```python
# torch.unsqueeze() adds a single dimension to a target tensor at a specific dim(dimension)
print(f"Previous tensor: \n {x_squeezed}")
print(f"Previous shape: {x_squeezed.shape} \n")

# Add an extra dimension with unsqueeze() to the end of the shape
x_unsqueezed = x_squeezed.unsqueeze(dim=0)
print(f"New tensor: \n {x_unsqueezed}")
print(f"New shape: {x_unsqueezed.shape}")
```

### Permute

Returns a view of the original tensor `input` with its dimensions permuted.

```python
# torch.permute() rearranges the dimensions of a target tensor 
# in a specified order and returns a view of the original tensor 
# with the same data but different shape.

x_original = torch.rand(size=(224, 224, 3)) # height, width, color channels

# Permute the original tensor to rearrange the axis(or dim) order
x_permuted = x_original.permute(2, 0, 1) # shift axis 0 to 1, axis 1 to 2, axis 2 to 0

print(f"Original shape: {x_original.shape}")
print(f"Permuted shape: {x_permuted.shape}")
```

## Indexing (Selecting data from tensors)

Indexing with PyTorch is similar to indexing with NumPy.

```python
# Create a tensor
import torch
x = torch.arange(1, 10).reshape(1, 3, 3)
x, x.shape
```

```python
# Let's index on our new tensor
x[0] # get the first (and only) item on the 0th axis
```

```python
# Let's index on the middle bracket (dim = 1)
x[0][0]
```

```python
#Let's index on the most inner bracket (dim = 2)
x[0][1][1]
```

```python
# You can also use ":" to select "all" of the target dimension
print(f"{x[:, 0]}") 

# Get the value of 0th and 1st dimensions but only index 1 of 2nd dimension
print(f"{x[:,:, 1]}")

# Get all values of the 0th dimension but only index 1 of 1st and 2nd dimensions
print(f"{x[:, 1, 1]}")

# Get index 0 of 0th and 1st dimensions but all values of 2nd dimension
print(f"{x[0, 0, :]}")
```

```python
# Index on x to return 9
# Note: 
# Indexing with integers returns a scalar (removes dimensions), 
# while slicing (:) returns a vector (preserves dimensions).
print(x[0, 2, 2])

# Index on x to return 3, 6, 9
print(x[:, :, 2 ])
```

## PyTorch tensor & NumPy

NumPy is a popular scientific Python numerical computing library.

And because of this, PyTorch has functionality to interact with it.

* Data in NumPy, want in PyTorch tensor -> `torch.from_numpy(ndarray)`
* PyTorch tensor -> NumPy -> `Torch.Tensor.Numpy()`



```python
# NumPy array to Pytorch tensor
import numpy as np
import torch

array = np.arange(1.0, 8.0)
# tensor = torch.from_numpy(array, type = torch.float32) error
# tensor = torch.from_numpy(array).type(torch.float32) right

tensor = torch.from_numpy(array)
# warning: when converting from numpy -> pytorch, 
# pytorch reflects numpy's default datatype of float64 
# unless specified otherwise
array, tensor
```

```python
# Change the value of array, what will this do to `tensor`?
array = array + 1
array, tensor
```

```python
# Tensor to NumPy array
tensor = torch.ones(7)
numpy_tensor = tensor.numpy()
tensor, numpy_tensor
```

```python
# Change the tensor, what will this do to `numpy_tensor`?
tensor = tensor + 1
tensor, numpy_tensor
```

## Reproducibility (trying to take random out of random)

In short how a neural network learns:

`start with random numbers -> tensor operations -> update random numbers to try and make them better representation of the data -> again -> again -> again...`

To reduce the randomness in neural networks and PyTorch comes the concept of a **random seed**.

Essentially what the random seed does is "flavour" the randomness.

```python
import torch

# Create two random tensors
random_tensor_A = torch.rand(3, 4)
random_tensor_B = torch.rand(3, 4)

print(f"Random tensor A: \n {random_tensor_A} \n")
print(f"Random tensor B: \n {random_tensor_B} \n")
print(random_tensor_A == random_tensor_B)
```

```python
# Let's make some random but reproducible tensors
import torch

# Set the random seed
RANDOM_SEED = 42

torch.manual_seed(RANDOM_SEED) 
random_tensor_C = torch.rand(3, 4)

torch.manual_seed(RANDOM_SEED) 
random_tensor_D = torch.rand(3, 4)

print(f"Random tensor C: \n {random_tensor_C} \n")
print(f"Random tensor D: \n {random_tensor_D} \n")

print(random_tensor_C == random_tensor_D)
```
